## 텍스트 데이터 전처리

- `[ ] : 일치시킬 문자 세트의 패턴
- [가나다] : 가 or 나 or 다 중에 하나를 포함하고 있는지
- [가-힣] : 한글 가부터 힣까의 문자 중 하나를 포함하고 있는지
- [0-9] : 0~9까지의 숫자 중 하나를 포함하고 있는지
- [^0-9] : 숫자를 포함하고 있지 않음
- [^가-힣] : 한글이 포함되어 있지 않음
- [가-힣+] : 한글이 하나 이상 포함되는지

|  <center>클래스</center> |  <center>표현식</center> |  <center>설명</center> |
|:--------|:--------:|--------:|
|**\d** | <center>[0-9] </center> |십진수와 일치 |
|**\D** | <center>[^0-9] </center> |숫자가 아닌 문자와 일치 |
|**\s** | <center>[\t\n\r\f\v] </center> |공백 문자와 일치|
|**\S** | <center>[^\t\n\r\f\v] </center> |공백이 아닌 문자와 일치|
|**\w** | <center>[a-zA-Z0-9_] </center> |모든 숫자 영문자와 일치|
|**\W** | <center>[^a-zA-Z0-9_] </center> |영문자, 숫자가 아닌 것과 일치|

- 표현식 설명

|  <center>표현식</center> |  <center>설명</center> |
|:--------|:--------:|
|**^X** | <center>문자열의 시작을 표현하며 x문자로시작 </center> |
|**x$** | <center> 문자열의 종료를 표현하며 x문자로 종료 </center> |
|**.x** | <center> 임의의 한 문자의 자리수를 표현하며 문자열이 x로 끝남 </center> |

|**(x)(y)** | <center> 그룹들의 집합을 표현하며 순서대로 번호를 부여하여 x,y는 각 그룹의 데이터로 관리 </center> |
|**(x)(?:y)** | <center> 그룹들의 집합에 대한 예외를 표현하며 그룹 집합으로 관리되지 않음 </center> |

### 편익시설

```python
df['공원보유시설(편익시설)'].value_counts()
```

```python
def get_string(text):
    if pd.isna(text):
        return text
    
    text = re.sub("[\d\(nan)\.'\-\[\]\/]","", text)
    text = re.sub(" ","", text)
    text = re.sub(",{2,}",",",text)
    return text
```

```python
get_string("[화장실, nan,4-, 55,/, nan]")
```

```python
get_string(np.nan)
```

```python
df['편익시설'] = df['공원보유시설(편익시설)'].map(get_string)
df['편익시설'].value_counts().tail()
```

```python
# 문자열을 연결해주는 join으로 편익시설을 연결해주는 방법이 있음
# 결측치가 있으면 결측치(np.nan)는 숫자형태(float)이기 때문에 오류가 발생
# 따라서 결측치를 제외한 값만 가져와서 연결

fac = ",".join(df.loc[df['편익시설'].notnull(),"편익시설"])
fac = get_string(fac)
fac[:100]
```

## 특정정보 마스킹

### 전화번호 마스킹

```python
# re.sub("규칙", "패턴"," 데이터")
# 규칙
# (\d{3}) 은 숫자 3자리를 의미
# (\d{1,2}) 는 숫자 1-2자리를 의미
# 패턴
# r"\1-\2**-**\5" 의 "\숫자" 는 앞에서 () 안에 정의한 값의 순서를 의미, 변수의 순서
# \3, \4는 쓰지않고 *로 대체
# r은 raw를 의미
```

```python
re.sub("(\d{3})-(\d{2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5"," 010-1234-5678")
```

```python
# 가운데가 3자리일때 번호 마스킹
re.sub("(\d{3})-(\d{1,2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5","010-123-5678")
```

```python
# 가운데가 2자리일때 번호 마스킹
re.sub("(\d{3})-(\d{1,2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5","010-12-5678")
```

```python
# 국가 번호가 들어갈때 확인하기 +82-10-1234-5678
# -가 들어가거나 여러 예외사항이 있을 때 마스킹되지 않을 수 있음, 함수를 만들어서 해결
re.sub("(\d{3})-(\d{2})(\d{2})-(\d{2})(\d{2})", r"\1-\2**-**\5", "+82-10-1234-5678")
```

```python
# 함수를 통해 가운데 들어오는 번호를 마스킹 하도록 처리
# 국제번호 등 다양한 형식의 번호가 들어왔을때
# -를 기준으로 뒤에서 두 번째 항목을 마스킹 처리 하되,
# 마지막 자리는 앞 두자리만 마스킹

phone = "+82-10-1234-5678"

def mask_phone_number(phone):
    if pd.isnull(phone):
        return phone
    
    num = phone.split("-")
    num_len = len(num) // 2
    
    num[num_len] = re.sub("\w","*", num[num_len])
    num[-1] = re.sub("(\d{2})(\d{2})",r"**\2", num[-1])
    
    return "-".join(num)

mask_phone_number(phone)
```

```python
# 결측치가 들어갔을때 오류가 날 수 있으므로 결측치를 체크
mask_phone_number(np.nan)
```

```python
df['전화번호(전처리)'] = df['전화번호'].map(mask_phone_number)
df[["공원명", "전화번호", "전화번호(전처리)"]].sample(10)
```
